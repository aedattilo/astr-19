```python
import matplotlib.pyplot as plt
from matplotlib import patches
import matplotlib.cm as cm
import numpy as np
```

```python
fig = plt.figure(figsize=(6,6))

# add a single subplot to the figure
# with empty x and y ticklabels
ax = fig.add_subplot(111, xticks=[], yticks=[])

# set the x and y limits
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

# blue rectangle
rectangle = patches.Rectangle((0.05, 0.05), 0.4, 0.4, edgecolor='None', facecolor='blue', alpha=0.5)
ax.add_patch(rectangle)

# red circle
circle = patches.Circle((0.6, 0.6), 0.3, edgecolor="None", facecolor='red', alpha=0.5)
ax.add_patch(circle)

# green ellipse
ellipse = patches.Ellipse((0.8, 0.8), 0.3, 0.4, edgecolor="None", facecolor='green', alpha=0.5)
ax.add_patch(ellipse)

plt.show()
```

# colorful mountains

```python
x = np.linspace(0, 2*np.pi, num=100)

#set the number of mountain ranges
nmr = 12
# set the number of peaks in each range
npeak = 12
```

```python
fig = plt.figure(figsize=(5, 10))

# get the turbo color map
cmap_turbo = cm.get_cmap('plasma')

# add a single subplot to the figure
# with empty x and y ticklabels
ax = fig.add_subplot(111, xticks=[], yticks=[])
ax.set_axis_off()

# set the x and y limits
ax.set_xlim([0, 2*np.pi])
ax.set_ylim([0, 15.])

for i in range(nmr):
    # set the color of this mountain rainge
    mr_color = cmap_turbo(float(i)/float(npeak))

    # create a randomized set of frequencies
    freq = 3.0*np.random.rand(npeak,) + 1.0

    # create a set of phase shifts
    phi = 2.0*np.pi*np.random.rand(npeak,)

    # create a set of amplitudes
    amp = np.random.rand(npeak) + 0.05

    # set an offset for that they all move
    yoff = 1*float(i)
    y = np.zeros_like(x)

    # for every peak
    for j in range(npeak):
        y += amp[j]*np.sin(freq[j]*x + phi[j])

    y = np.fabs(y) + yoff
    plt.fill_between(x, y, yoff-1, zorder=nmr-i, color=mr_color, cmap='plasma', alpha=1-0.005*i)

plt.savefig('rainbow_mountains.png', dpi=300)
plt.show()
```
