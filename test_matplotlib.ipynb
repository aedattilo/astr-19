```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 5, 0.1) # create x=[0..5] in 0.1 increments
y = np.sin(x)            # y = sin(x)

plt.figure()             # instantiate your figure
plt.plot(x, y)           # plot data
plt.xlabel('x')          # label x-axis
plt.ylabel('sin(x)')     # label y-axis
plt.savefig("sinx.png", bbox_inches='tight', dpi=200)
plt.show()               # display the plot
```

```python
# plot multiple datasets on one plot

x = np.arange(1, 19, 0.4)
y1 = np.log10(x)
y2 = 0.01 * x**2
y3 = 0.9 * np.sin(x)

plt.figure(facecolor='oldlace')
plt.plot(x, y1, 'r-', label='y1')
plt.plot(x, y2, 'b^', label='y2')
plt.plot(x, y3, 'go', label='y3')
plt.plot(x, y1+y2+y3, 'm+', label='y1+y2+y3')
plt.legend(loc=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = np.linspace(0, 9, num=10) + 2*np.random.randn(10)
y = np.linspace(0, 9, num=10) + 2*np.random.randn(10)

# generate fake errors for the data
xerr = 2*np.random.random(10)
yerr = 2*np.random.random(10)

plt.figure()
plt.errorbar(x, y, xerr=xerr, yerr=yerr, fmt='m.')
# plt.xlabel('Daily high temperature')
# plt.ylabel("Hours fan runs")
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, num=100)
print(x[-1], 2*np.pi)

y = np.sin(x)
z = np.cos(x)
w = np.sin(4*x)
v = np.cos(4*x)
```

```python
fig, axs = plt.subplots(1, 2)

axs[0].plot(x, y)
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].set_title(r'$\sin(x)$')

axs[1].plot(x, z)
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].set_title(r'$\cos(x)$')

fig.subplots_adjust(wspace=0.4)
axs[0].set_aspect('equal')
axs[1].set_aspect(np.pi)
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))

plt.plot(x, y, label=r'$y = \sin(x)$')
plt.plot(x, z, label=r'$y = \cos(x)$')
plt.plot(x, w, label=r'$y = \sin(4x)$')
plt.plot(x, v, label=r'$y = \cos(4x)$')

plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.xlim(0, 2*np.pi)
plt.ylim(-1.2, 1.2)
plt.legend(loc=1, framealpha=0.95)

plt.gca().set_aspect(np.pi/1.2)

plt.show()
```
