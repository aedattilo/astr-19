```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
```

```python
x = np.linspace(0, 1*np.pi, num=200)
y = np.linspace(0, 2*np.pi, num=200)
xg, yg = np.meshgrid(x, y, indexing='ij')
z = -np.sin(2*xg)*np.sin(yg)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)

im = plt.pcolormesh(xg, yg, z, cmap=cm.turbo)
cbar = plt.colorbar(im, ticks=[-0.5, 0.0, 0.5])
cbar.ax.get_yaxis().labelpad=15
cbar.ax.set_ylabel('Z Value', rotation=270, fontsize=14)

plt.xlabel("X Label", fontsize=14)
plt.ylabel("Y Label", fontsize=14)
plt.show()
```

```python
%matplotlib inline
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(xg, yg, z, rstride=5, cstride=5, linewidth=0.5)
ax.view_init(20)
ax.set_xlabel("X Label", fontsize=14)
ax.set_ylabel("Y Label", fontsize=14)
ax.set_zlabel("Z Label", fontsize=14)

plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(xg, yg, z, rstride=4, cstride=4, cmap=cm.coolwarm, alpha=0.7)
fig.colorbar(surf, shrink=0.5, label="Z label")

ax.plot_wireframe(xg, yg, z, rstride=10, cstride=10, linewidth=0.5, color='gray')
cset = ax.contourf(xg, yg, z, zdir='z', cmap=cm.coolwarm, offset=-1)

ax.view_init(10)
ax.set_xlabel("X Label", fontsize=14)
ax.set_ylabel("Y Label", fontsize=14)

# ax.w_zaxis.line.set_lw(0.)
ax.set_zticks([])

plt.show()
```
