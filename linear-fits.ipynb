```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(119)

npoints = 50

x = np.linspace(0, 10, num=npoints)

# truth
m = 2.
b = 1.
sigma = 2.

y = m*x + b + np.random.normal(0, sigma, size=npoints)
y_err = np.full(npoints, sigma)

plt.figure()
plt.plot(x, m*x+b)
plt.errorbar(x, y, yerr=y_err, fmt='r.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Method #1, polyfit()

```python
m_fit, b_fit = np.poly1d(np.polyfit(x, y, 1, w=1./y_err))
print(m_fit, b_fit)

y_fit = m_fit*x + b_fit

plt.figure()
plt.plot(x, m*x+b, c='gray', linestyle="--", label="truth")
plt.errorbar(x, y, yerr=y_err, fmt='r.', label='data')
plt.plot(x, y_fit, label="polyfit()")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=2)
plt.show()
```
